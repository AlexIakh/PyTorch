```python
torch.manual_seed(1234)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f'Device: {device}')

tens_6 = torch.rand(2, 3).to(device)
tens_61 = torch.rand(2, 3).to(device)

tens_6, tens_61
```

```python
mul_tensors = torch.matmul(tens_6, tens_61.T)
mul_tensors, mul_tensors.shape
```

```python
max = mul_tensors.max()
min = mul_tensors.min()
max, min
```

```python
arg_max = torch.argmax(mul_tensors)
arg_min = torch.argmin(mul_tensors)
arg_max, arg_min
```
